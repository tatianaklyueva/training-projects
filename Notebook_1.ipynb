### Заполнение пропусков

В некоторых столбцах есть приопущенные значения. Оценим их количество. 

```python
df.isna().sum() # подсчёт пропусков
```

Итак, пропуски есть в двух столбцах. Один из них — `days_employed`, другой - `total_income`. Интересно, что пропусков одинаковое количество. 
Проанализируем эти столбцы. Запросим первые 10 строк с пропущенными значениями.

```python
# просмотр 10 первых строк с пропущенными значениями
display(df[df['total_income'].isna()].head(10))
```

Пропущенные значения не идут подряд, они не связаны ни с особенностями клиентов (возрастом, занятостью),ни с типом кредита. Однако пропуски в столбце `days_employed` соответствуют пропускам в столбце `total_income`. Проверим, так ли это во всей таблице.

```python
# Количество строк, где одновременно пропущенны значения в столбцах 'days_employed' и 'total_income'
df_isna = len(df.loc[df['days_employed'].isna() & df['total_income'].isna()])
print(f'Количество строк, где пропущенны значения одновременно в обоих столбцах {df_isna}')

# Доля этого количества от общего количества наблюдений
na_share_2 = df.isna().mean()
print(f'Доля строк с пропущенными значениями от общего числа строк в таблице составляет {na_share_2}')
```

Пропущенные значения составляют примерно 10% в каждом из столбцов с пропусками. Это немало.
Только столбцы `days_employed`и `total_income` в датафрейме имеют формат `float64`. Совпадение пропусков в этих столбцах говорит о том, что может быть допущена техническая ошибка при выгрузке или обработке данных. 
В реальном проекте необходимо было бы обратиться к источнику данных, найти и исправить техническую ошибку.

У нас такой возможности нет, поэтому постараемся заполнить отсутствующие данные типичными значениями выборки. 
Начнем со столбца `total_income`. Для заполнения пропусков можно использовать типичные значения выборки - среднее арифметическое или медиану. Рассчитаем оба этих показателя.

```python
mean_total_income = df['total_income'].mean() # расчет среднего арифметического по столбцу
median_total_income = df['total_income'].median()  # расчет медианы по столбцу

print(f'Среднее арифметическое значение по столбцу \'Ежемесячный доход\' - {mean_total_income:.2f}' )
print(f'Медианное значение по столбцу \'Ежемесячный доход\' - {median_total_income:.2f}' )
```

Мы видим, что среднее арифметическое значение значительно выше медианного. Это объясняется тем, что медианное значение менее чувствительно к 'выбросам' - очень большим и очень маленьким значениям в выборке, которые могут сдвигать среднее. В нашем случае малое количество заемщиков с очень высокими доходами влияет на среднее арифметическое, повышая его. На медиану 'выбросы' практически не влияют.

Поэтому лучшим решением будет выбрать для замены медианное значение. 

```python
# Заполнение пропущенных значений с помощью метода fillna()
df['total_income'] = df['total_income'].fillna(df.groupby('income_type')['total_income'].transform('median'))# Проверка на отсутствие пропущенных значений
total_income_isna = df['total_income'].isna().sum()
print(f'Количество строк, где пропущенны значения в столбцe \'Ежемесячный доход\' - {total_income_isna}')
```

Мы заменили все пропущенные значения в столбце `total_income` на медианные. Перейдем к столбцу `days_employed`

```python
# Замена значений стажа медианными по типу дохода. 
df.loc[df['income_type'] == 'компаньон','days_employed'] = df.loc[df['income_type'] == 'компаньон','days_employed'].fillna(median_comp)
df.loc[df['income_type'] == 'предприниматель','days_employed'] = df.loc[df['income_type'] == 'предприниматель','days_employed'].fillna(median_comp)
df.loc[df['income_type'] == 'пенсионер','days_employed'] = df.loc[df['income_type'] == 'пенсионер','days_employed'].fillna(median_elder)
df.loc[df['income_type'] == 'сотрудник','days_employed'] = df.loc[df['income_type'] == 'сотрудник','days_employed'].fillna(median_empl)
df.loc[df['income_type'] == 'госслужащий','days_employed'] = df.loc[df['income_type'] == 'госслужащий','days_employed'].fillna(median_state)
```

### Категоризация дохода

На основании диапазонов, указанных ниже, создадим столбец total_income_category с категориями:
0–30000 — 'E';
30001–50000 — 'D';
50001–200000 — 'C';
200001–1000000 — 'B';
1000001 и выше — 'A'.
Например, кредитополучателю с доходом 25000 нужно назначить категорию 'E', а клиенту, получающему 235000, — 'B'.

```python
def choose_category(x):
    if 0 < x <= 30_000:
        return 'E'
    if x <= 50_000:
        return 'D'
    if x <= 200_000:
        return 'C'
    if x <= 1_000_000:
        return 'B'
    if x >= 1_000_001:
        return 'A'
    
df['total_income_category'] = df['total_income'].transform(choose_category)
df.head()
```

### Категоризация целей кредита

Изучим данные в столбце purpose и определим, какие подстроки помогут нам правильно определить категорию кредита.

```python
df['purpose'].value_counts()
```

Создадим функцию, которая на основании данных из столбца purpose сформирует новый столбец purpose_category, в который войдут следующие категории:
* 'операции с автомобилем' (подстрока 'автомобил'),
* 'операции с недвижимостью' (подстроки 'недвижим', 'жиль')
* 'проведение свадьбы',(подстрока 'свадьб')
* 'получение образования' (подстока 'образован').

Например, если в столбце purpose находится подстрока 'на покупку автомобиля', то в столбце purpose_category должна появиться строка 'операции с автомобилем'.
Используем собственную функцию и метод apply(). 

```python
def label_purpose_category(row):
    if 'автомобил' in row:
        return 'операции с автомобилем'
    if 'недвижим' in row or 'жиль' in row:
        return 'операции с недвижимостью'
    if 'свадьб' in row:
        return 'проведение свадьбы'
    if 'образован' in row:
        return 'получение образования'
    return 'другое'
df['purpose_category'] = df['purpose'].apply(label_purpose_category)
df.head()
```

Проверим, что все наши данные корректно классифицированы, категория 'другое' не встречается.

```python
df[df['purpose_category']=='другое']
```
